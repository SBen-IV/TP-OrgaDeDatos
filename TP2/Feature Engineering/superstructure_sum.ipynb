```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv('../Train/train_set_geo_level_sum.csv')

train.head()
```

```python
def _sum_cols(x, cols, i, n):
    if i == n:
        return 0
    return x[cols[i]] + _sum_cols(x, cols, i+1, n)

def sum_cols(x, cols):
    return _sum_cols(x, cols, 0, len(cols))
```

```python
cols = list(train.columns)

superstructure = []
for col in cols:
    if col.startswith("has_superstructure"):
        superstructure.append(col)
        
superstructure
```

```python
pos = len(cols) - 1
train.insert(pos, "superstructure_sum", sum_cols(train, superstructure))
```

```python
train['superstructure_sum'].value_counts()
```

```python
train.to_csv('../Train/train_set_geo_superstructure.csv', index=False)
```
