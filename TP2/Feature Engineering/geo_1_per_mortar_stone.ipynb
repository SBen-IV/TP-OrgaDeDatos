```python
import pandas as pd
import numpy as np
```

## Train

geo level sum

```python
train = pd.read_csv('../Train/train_set_geo_level_sum.csv')

train.head()
```

```python
pos = len(train.columns) - 1

train.insert(pos, "geo_level_1_per_mud_mortar_stone", train['geo_level_1_id'] * train['has_superstructure_mud_mortar_stone'])
pos +=1
train.insert(pos, "geo_level_1_sum_per_mud_mortar_stone", train['geo_level_1_id_sum'] * train['has_superstructure_mud_mortar_stone'])

train.head()
```

```python
train.to_csv('../Train/train_set_geo_1_per_mortar_stone.csv', index=False)
```
