```python
import pandas as pd
import numpy as np
```

## Train

float

```python
train = pd.read_csv('../Train/train_set_geo_level_sum.csv')

train.head()
```

```python
train['base_percentage'] = train['area_percentage'] / train['height_percentage']

train['base_percentage'].value_counts()
```

```python
cols = list(train.columns)
cols.remove('damage_grade')
cols.append('damage_grade')

train = train[cols]

train
```

```python
train.to_csv('../Train/train_set_geo_level_sum_base.csv', index=False)
```
