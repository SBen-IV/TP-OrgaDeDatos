```python
import pandas as pd
```

## Train

```python
train = pd.read_csv('../Train/train_set_sin_building_id.csv')

train.head()
```

### Categorical columns

```python
train['legal_ownership_status'] = train['legal_ownership_status'].replace({"a": 0, "r": 1, "v": 2, "w": 3})
train['plan_configuration'] = train['plan_configuration'].replace({"a": 0, "c": 1, "d": 2, "f": 3, "m": 4, "n": 5, "o": 6, "q": 7, "s": 8, "u": 9})
train['position'] = train['position'].replace({"j": 0, "o": 1, "s": 2, "t": 3})
train['other_floor_type'] = train['other_floor_type'].replace({"j": 0, "q": 1, "s": 2, "x": 3})
train['ground_floor_type'] = train['ground_floor_type'].replace({"f": 0, "m": 1, "v": 2, "x": 3, "z": 4})
train['roof_type'] = train['roof_type'].replace({"n": 0, "q": 1, "x": 2})
train['foundation_type'] = train['foundation_type'].replace({"h": 0, "i": 1, "r": 2, "u": 3, "w": 4})
train['land_surface_condition'] = train['land_surface_condition'].replace({"n": 0, "o": 1, "t": 2})
```

```python
train.head()
```

### Has superstructure

```python
train['has_superstructure_general'] = train['has_superstructure_adobe_mud']
del train['has_superstructure_adobe_mud']

train['has_superstructure_general'] = train['has_superstructure_general'] + train['has_superstructure_mud_mortar_stone'] *2
del train['has_superstructure_mud_mortar_stone']

train['has_superstructure_general'] = train['has_superstructure_general'] + train['has_superstructure_stone_flag'] *3
del train['has_superstructure_stone_flag'] 

train['has_superstructure_general'] = train['has_superstructure_general'] + train['has_superstructure_cement_mortar_stone'] *4
del train['has_superstructure_cement_mortar_stone']

train['has_superstructure_general'] = train['has_superstructure_general'] + train['has_superstructure_mud_mortar_brick'] *5
del train['has_superstructure_mud_mortar_brick']

train['has_superstructure_general'] = train['has_superstructure_general'] + train['has_superstructure_cement_mortar_brick'] *6
del train['has_superstructure_cement_mortar_brick']

train['has_superstructure_general'] = train['has_superstructure_general'] + train['has_superstructure_timber'] *7
del train['has_superstructure_timber']

train['has_superstructure_general'] = train['has_superstructure_general'] + train['has_superstructure_bamboo'] *8
del train['has_superstructure_bamboo']

train['has_superstructure_general'] = train['has_superstructure_general'] + train['has_superstructure_rc_engineered'] *9
del train['has_superstructure_rc_engineered']

train['has_superstructure_general'] = train['has_superstructure_general'] + train['has_superstructure_rc_non_engineered'] *10
del train['has_superstructure_rc_non_engineered']

train['has_superstructure_general'] = train['has_superstructure_general'] + train['has_superstructure_other'] *11
del train['has_superstructure_other']

train
```

### Has secondary use

```python
train['has_secondary_general'] = train['has_secondary_use']
del train['has_secondary_use']

train['has_secondary_general'] = train['has_secondary_general'] + train['has_secondary_use_agriculture'] *2
del train['has_secondary_use_agriculture']

train['has_secondary_general'] = train['has_secondary_general'] + train['has_secondary_use_hotel'] *3
del train['has_secondary_use_hotel'] 

train['has_secondary_general'] = train['has_secondary_general'] + train['has_secondary_use_rental'] *4
del train['has_secondary_use_rental']

train['has_secondary_general'] = train['has_secondary_general'] + train['has_secondary_use_institution'] *5
del train['has_secondary_use_institution']

train['has_secondary_general'] = train['has_secondary_general'] + train['has_secondary_use_school'] *6
del train['has_secondary_use_school']

train['has_secondary_general'] = train['has_secondary_general'] + train['has_secondary_use_industry'] *7
del train['has_secondary_use_industry']

train['has_secondary_general'] = train['has_secondary_general'] + train['has_secondary_use_health_post'] *8
del train['has_secondary_use_health_post']

train['has_secondary_general'] = train['has_secondary_general'] + train['has_secondary_use_gov_office'] *9
del train['has_secondary_use_gov_office']

train['has_secondary_general'] = train['has_secondary_general'] + train['has_secondary_use_use_police'] *10
del train['has_secondary_use_use_police']

train['has_secondary_general'] = train['has_secondary_general'] + train['has_secondary_use_other'] *11
del train['has_secondary_use_other']

train
```

### count floors, area y height percentage

```python
train['espacio_gral_edad'] = train['count_floors_pre_eq'] * train['area_percentage'] * train['height_percentage'] * train['age']
```

### Geo level id

```python
train['geo_level_total'] = train['geo_level_1_id'] * train['geo_level_2_id'] * train['geo_level_3_id']
del train['geo_level_1_id']
del train['geo_level_2_id']
del train['geo_level_3_id']
```

```python
cols = list(train.columns)
cols.remove('damage_grade')
cols.append('damage_grade')

train = train[cols]

train.head()
```

```python
train.to_csv('../Train/train_set_idea.csv', index=False)
```
