```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv('../Train/train_set_geo_level_sum.csv')

train.head()
```

```python
pos = len(train.columns) - 1

train.insert(pos, "geo_level_sum", train['geo_level_1_id'] + train['geo_level_2_id'] + train['geo_level_3_id'])

train.head()
```

```python
train.insert(pos, "geo_level_multiply", train['geo_level_1_id'] * train['geo_level_2_id'] * train['geo_level_3_id'])

train.head()
```

```python
train.to_csv('../Train/train_set_geo_levels_sum_mult.csv', index=False)
```
