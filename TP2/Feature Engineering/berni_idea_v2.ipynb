```python
import pandas as pd
```

## Train

```python
train = pd.read_csv('../Train/train_set_sin_building_id.csv')

train.head()
```

```python
train['legal_ownership_status'] = train['legal_ownership_status'].replace({"a": 0, "r": 1, "v": 2, "w": 3})
train['plan_configuration'] = train['plan_configuration'].replace({"a": 0, "c": 1, "d": 2, "f": 3, "m": 4, "n": 5, "o": 6, "q": 7, "s": 8, "u": 9})
train['position'] = train['position'].replace({"j": 0, "o": 1, "s": 2, "t": 3})
train['other_floor_type'] = train['other_floor_type'].replace({"j": 0, "q": 1, "s": 2, "x": 3})
train['ground_floor_type'] = train['ground_floor_type'].replace({"f": 0, "m": 1, "v": 2, "x": 3, "z": 4})
train['roof_type'] = train['roof_type'].replace({"n": 0, "q": 1, "x": 2})
train['foundation_type'] = train['foundation_type'].replace({"h": 0, "i": 1, "r": 2, "u": 3, "w": 4})
train['land_surface_condition'] = train['land_surface_condition'].replace({"n": 0, "o": 1, "t": 2})
```

```python
train.head()
```

```python
train.to_csv('../Train/train_set_idea_v2.csv', index=False)
```
