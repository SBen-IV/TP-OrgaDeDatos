```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
```

## Train

geo level sum

```python
train = pd.read_csv('../Train/train_set_geo_level_sum.csv')

train.head()
```

```python
X, y = train.iloc[:,:-1], train.iloc[:,-1]
```

```python
rf_model = RandomForestClassifier(n_estimators=15, max_depth=3, max_features='log2', random_state=123)
rf_model.fit(X, y)
```

```python
leaf_index = rf_model.apply(X)

leaf_index
```

```python
leaf_index_df = pd.DataFrame(leaf_index)

leaf_index_df
```

```python
cols = list(leaf_index_df.columns)

new_cols = ["leaf_index_"+str(n) for n in cols]

new_cols
```

```python
leaf_index_df.columns = new_cols

leaf_index_df
```

```python
result = train.join(leaf_index_df)

result
```

```python
cols = list(result.columns)
cols.remove('damage_grade')
cols.append('damage_grade')

result = result[cols]

result
```

```python
result.to_csv('../Train/train_set_geo_sum_base_rf_leaf_index.csv', index=False)
```
