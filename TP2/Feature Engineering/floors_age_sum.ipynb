```python
import pandas as pd
import numpy as np
```

## Función

```python
def add_sum_cols(df, cols):
    for col in cols:
        values = pd.get_dummies(df[col]).sum().to_dict()

        new_col_name = col + "_sum"

        df[new_col_name] = df[col].map(values)
```

## Train

```python
train = pd.read_csv('../Train/train_set_geo_level_sum.csv')

train.head()
```

```python
cols = ['count_floors_pre_eq', 'age']
```

```python
add_sum_cols(train, cols)

train.head()
```
