```python
import pandas as pd
```

```python
def remove_secondary_use(cols):
    new_cols = []
    for col in cols:
        if not col.startswith('has_secondary'):
            new_cols.append(col)
    
    return new_cols
```

## Train

One Hot Encoding

```python
train = pd.read_csv('../Train/train_set_one_hot_encoding.csv')

train.head()
```

```python
new_cols = remove_secondary_use(list(train.columns))
train = train[new_cols]

train.head()
```

```python
train.to_csv('../Train/train_set_one_hot_encoding_sin_secondary_use.csv', index=False)
```

Binary Encoding

```python
train = pd.read_csv('../Train/train_set_binary_encoding.csv')

train.head()
```

```python
new_cols = remove_secondary_use(list(train.columns))
train = train[new_cols]

train.head()
```

```python
train.to_csv('../Train/train_set_binary_encoding_sin_secondary_use.csv', index=False)
```
