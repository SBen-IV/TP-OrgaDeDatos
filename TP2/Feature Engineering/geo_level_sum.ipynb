```python
import pandas as pd
import numpy as np
```

## Funciones

```python
def add_sum_cols(df, cols):
    for col in cols:
        values = pd.get_dummies(df[col]).sum().to_dict()

        new_col_name = col + "_sum"

        df[new_col_name] = df[col].map(values)
```

## Train

```python
train = pd.read_csv('../Train/train_set_sin_secondary_use.csv')

train.head()
```

```python
main_cols = ['geo_level_1_id', 'geo_level_2_id', 'geo_level_3_id']
```

```python
add_sum_cols(train, main_cols)
```
