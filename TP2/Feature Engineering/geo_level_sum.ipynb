```python
import pandas as pd
import numpy as np
import seaborn as sns
```

## Funciones

```python
def add_sum_cols(df, cols):
    for col in cols:
        values = pd.get_dummies(df[col]).sum().to_dict()

        new_col_name = col + "_sum"

        df[new_col_name] = df[col].map(values)
```

## Train

sin secondary use

```python
train = pd.read_csv('../Train/train_set_sin_secondary_use.csv')

train.head()
```

```python
main_cols = ['geo_level_1_id', 'geo_level_2_id', 'geo_level_3_id']
```

```python
add_sum_cols(train, main_cols)
```

```python
train
```

```python
cols = list(train.columns)
cols.remove('damage_grade')
cols.append('damage_grade')

train = train[cols]

train
```

```python
chosen_cols = ['geo_level_1_id', 'geo_level_2_id', 'geo_level_3_id', 'geo_level_1_id_sum', 'geo_level_2_id_sum', 'geo_level_3_id_sum',\
              'damage_grade']
sns.pairplot(train[chosen_cols], hue='damage_grade')
```

```python
train.to_csv('../Train/train_set_geo_level_sum.csv', index=False)
```
