```python
import pandas as pd
```

## Train

```python
train_set = pd.read_csv('../Train/train_set_sin_building_id.csv')

train_set.head(10)
```

```python
columns = train_set.columns

for col in columns:
    if train_set[col].dtype == 'O':
        train_set.drop(col, axis='columns', inplace=True)

train_set.head()
```

```python
train_set.info()
```

```python
train_set.to_csv('../Train/train_set_sin_categoricos.csv', index=False)
```
