```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import f1_score
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
```

```python
train = pd.read_csv('../Train/train_set_geo_floors_age_sum.csv')

train.head()
```

```python
X, y = train.iloc[:,:-1], train.iloc[:,-1]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)
```

```python
xgb_model = xgb.XGBClassifier(n_estimators=5, max_depth=15, learning_rate=0.01, random_state=123, gamma=0, reg_lambda=0.4, verbosity=0)
xgb_model.fit(X_train, y_train)
preds = xgb_model.predict(X_test)
f1 = f1_score(y_test, preds, average='micro'),
print("F1-Mean Score %f" % (f1))
```

```python
xgb_model = xgb.XGBClassifier(n_estimators=30, max_depth=30, learning_rate=0.3, random_state=123, gamma=4.5, reg_lambda=3.5, verbosity=0)
xgb_model.fit(X_train, y_train)
preds = xgb_model.predict(X_test)
f1 = f1_score(y_test, preds, average='micro'),
print("F1-Mean Score %f" % (f1))
```

```python
params = {
    'n_estimators': [5, 15, 30],
    'max_depth': [15, 30],
    'learning_rate': [0.01, 0.1, 0.3],
    'random_state': [123],
    'gamma': [0, 1.5, 4.5],
    'reg_lambda': [0.4, 1.3, 3.5]
}
```

```python
opt = GridSearchCV(xgb.XGBClassifier(), params, cv=10)
opt.fit(X_train, y_train)
```

```python
opt.best_estimator_
```

```python
preds = opt.predict(X_test)
f1 = f1_score(y_test, preds, average='micro'),
print("F1-Mean Score %f" % (f1))
```
